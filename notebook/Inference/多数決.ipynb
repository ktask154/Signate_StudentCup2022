```python
import pandas as pd
import numpy as np
import os
```

```python
INPUT_DIR = '/content/drive/MyDrive/Competitions/Signate/Student Cup 2022/input/'
OUTPUT_DIR = '/content/drive/MyDrive/Competitions/Signate/Student Cup 2022/output/'
OUTPUT_SUB_DIR = os.path.join(OUTPUT_DIR,'Submission')
```

```python
sub10 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"7327submission10.csv"),header=None)
sub10.columns = ["id","sub10"]
sub12 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"7455submission12.csv"),header=None)
sub12.columns = ["id","sub12"]
sub16 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"7385submission16.csv"),header=None)
sub16.columns = ["id","sub16"]
sub20 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"7301submission20.csv"),header=None)
sub20.columns = ["id","sub20"]
sub21 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"7525submission21.csv"),header=None)
sub21.columns = ["id","sub21"]
sub22 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"0.7473submission22[2+3+5].csv"),header=None)
sub22.columns = ["id","sub22"]
sub24 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"7347submission24.csv"),header=None)
sub24.columns = ["id","sub24"]
sub26 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"8107submission26.csv"),header=None)
sub26.columns = ["id","sub26"]
sub27 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"7150submission27.csv"),header=None)
sub27.columns = ["id","sub27"]
sub28 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"7460submission28.csv"),header=None)
sub28.columns = ["id","sub28"]
sub30 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"???submission30.csv"),header=None)
sub30.columns = ["id","sub30"]
sub31 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"7429submission31.csv"),header=None)
sub31.columns = ["id","sub31"]
sub32 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"submission32_Stacking.csv"),header=None)
sub32.columns = ["id","sub32"]
sub33 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"submission33 ensemble[12+30+31].csv"),header=None)
sub33.columns = ["id","sub33"]
sub34 = pd.read_csv(os.path.join(OUTPUT_SUB_DIR,"submission34ensemble[12+31].csv"),header=None)
sub34.columns = ["id","sub34"]
```

```python
df = sub10.merge(sub12,how='left',on='id').merge(sub16,how='left',on='id').merge(sub20,how='left',on='id').merge(sub21,how='left',on='id').merge(sub22,how='left',on='id').merge(sub24,how='left',on='id').merge(sub26,how='left',on='id').merge(sub27,how='left',on='id').merge(sub28,how='left',on='id').merge(sub30,how='left',on='id').merge(sub31,how='left',on='id').merge(sub32,how='left',on='id').merge(sub33,how='left',on='id').merge(sub34,how='left',on='id')
```

```python
df
```

```python
df["label"] = df.loc[:,"sub10":].mode(axis=1)[0]
df["label"] = df["label"].astype("int")
df["relabel"] = df["label"]
df.loc[df.sub12==df.sub31,"relabel"] = df.sub31
df
```
